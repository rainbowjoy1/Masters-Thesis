```python
#RNN Merge

from functools import reduce

import pandas as pd
import pickle
```

```python
#2010
df_10 = pd.read_pickle("RNN_Test_Sentences_2010")
df_10 = df_10.drop_duplicates()
#2011
df_11 = pd.read_pickle("RNN_Test_Sentences_2011")
df_11 = df_11.drop_duplicates()
#2012
df_12 = pd.read_pickle("RNN_Test_Sentences_2012")
df_12 = df_12.drop_duplicates()
#2013
df_13 = pd.read_pickle("RNN_Test_Sentences_2013")
df_13 = df_13.drop_duplicates()
#2014
df_14 = pd.read_pickle("RNN_Test_Sentences_2014")
df_14 = df_14.drop_duplicates()
#2015
df_15 = pd.read_pickle("RNN_Test_Sentences_2015")
df_15 = df_15.drop_duplicates()
#2016
df_16 = pd.read_pickle("RNN_Test_Sentences_2016")
df_16 = df_16.drop_duplicates()
#2017
df_17 = pd.read_pickle("RNN_Test_Sentences_2017")
df_17 = df_17.drop_duplicates()
#2018
df_18 = pd.read_pickle("RNN_Test_Sentences_2018")
df_18 = df_18.drop_duplicates()
#2019
df_19 = pd.read_pickle("RNN_Test_Sentences_2019")
df_19 = df_19.drop_duplicates()
#2020
df_20 = pd.read_pickle("RNN_Test_Sentences_2020")
df_20 = df_20.drop_duplicates()
#2021
df_21 = pd.read_pickle("RNN_Test_Sentences_2021")
df_21 = df_21.drop_duplicates()
#2022
df_22 = pd.read_pickle("RNN_Test_Sentences_2022")
df_22 = df_22.drop_duplicates()
```

```python
df_list = [df_10, df_11, df_12, df_13, df_14, df_15, df_16, df_17, df_18, df_19, df_20, df_21, df_22]
dfs = [df.drop_duplicates()for df in df_list]
dfs = [df.drop(df.columns[[1,2,3,5,6,7,8]],axis = 1) for df in dfs]
dfs = [df.groupby(['feature','source']).mean()for df in dfs]
```

```python
data_merge = reduce(lambda left, right: pd.merge(left , right, on = ['feature','source'],how = "inner"), dfs)
```
